```r
library(tidyr)
top_10 <- top_n(da_final,10,whites)
```

```r
#top_10
```

```r
library(dplyr)
plotdata <- top_10 %>%
  group_by(STNAME) %>%
  summarize(mean_white = mean(whites))

ggplot(plotdata,aes(x = STNAME, y = mean_white)) +
  geom_bar(stat = "identity")
```

```r
top_10_b <- top_n(da_final,10,blacks)
#top_10_b
```

```r
library(dplyr)
plotdata <- top_10_b %>%
  group_by(STNAME) %>%
  summarize(mean_black = mean(blacks))

# plot mean salaries
ggplot(plotdata,aes(x = STNAME, y = mean_black)) +
  geom_bar(stat = "identity")
```

```r
top_10_a <- top_n(da_final,10,asians)
#top_10_a
```

```r
library(dplyr)
plotdata <- top_10_a %>%
  group_by(STNAME) %>%
  summarize(mean_asian = mean(asians))

# plot mean salaries
ggplot(plotdata,aes(x = STNAME, y = mean_asian)) +
  geom_bar(stat = "identity")
```

```r
top_10_i <- top_n(da_final,10,indians)
```

```r
library(dplyr)
plotdata <- top_10_i %>%
  group_by(STNAME) %>%
  summarize(mean_indians = mean(indians))

# plot mean salaries
ggplot(plotdata,aes(x = STNAME, y = mean_indians)) +
  geom_bar(stat = "identity")
```
